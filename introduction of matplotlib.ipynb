```python
heart_disease=pd.read_csv("heart-disease.csv")
heart_disease.head()
```

```python
#create a histogram of age
heart_disease["age"].plot.hist();
```

```python
heart_disease.plot.hist(figsize=(10,20),subplots=True);
```

```python
#which method should we use ?(pyplot vsmatplotlib OO method)
* when plotting something quickly ,okay to use the pyplot method
* when Plotting something more advanced ,use the OO method.
```

```python
heart_disease.head()
```

```python
Over_50=heart_disease[heart_disease["age"]>50]
Over_50.head()
```

```python
#Pyplot method
Over_50.plot(kind="scatter",
            x='age'
            ,y="chol"
            ,c="target");
```

```python
# OO method mixed with pyplot method
fix,ax=plt.subplots(figsize=(10,6))
Over_50.plot(kind='scatter',
            x='age',
            y='chol',
            c='target',
            ax=ax);
# ax.set_xlim([45,100]); setting up the limit of x axis
```

```python
# OO method with scratch
fig,ax=plt.subplots(figsize=(10,6))

#plot the data
scatter=ax.scatter(x=Over_50["age"],
                  y=Over_50["chol"],
                  c=Over_50["target"])
#customise the plot
ax.set(title="Heart disease and cholestrol levels",
    xlabel="Age",
      ylabel="cholestrol");
#Add a legend
ax.legend(*scatter.legend_elements(),title="Target");

# add a hortizonatal line
ax.axhline(Over_50["chol"].mean(),
          linestyle='--');
```

```python
# we are going to make combine part two dataframe
fig,(ax0,ax1)=plt.subplots(nrows=2,
                          ncols=1,
                          figsize=(10,10),
                           sharex=True)
# Add data to the 'ax0'
scatter=ax0.scatter(x=Over_50["age"],
                   y=Over_50["chol"],
                   c=Over_50["target"]);

#customise the set
ax0.set(title="heart disease and cholestrol level",
       ylabel="cholestrol")
# Add legend
ax0.legend(*scatter.legend_elements(),title="Target");

#Add a line
ax0.axhline(Over_50["chol"].mean(),
            linestyle="--");

#Adding data to the ax1
scatter=ax1.scatter(x=Over_50["age"],
                    y=Over_50["thalach"],
                    c=Over_50["target"])
ax1.set(title="heart disease and Heart rate",
       xlabel="Age",
       ylabel="Max heart rate");
# Add legend
ax1.legend(*scatter.legend_elements(),title="Target");

#Add a line
ax1.axhline(Over_50["thalach"].mean(),
            linestyle="--");
#Add title to figure
fig.suptitle("Heart Disease Analysis",fontsize=16,fontweight="bold");
```

```python
plt.style.use('seaborn-v0_8-bright');

# OO method with scratch
fig,ax=plt.subplots(figsize=(10,6))

#plot the data
scatter=ax.scatter(x=Over_50["age"],
                  y=Over_50["chol"],
                  c=Over_50["target"],
                  cmap="winter")# this changes the color scheme
#customise the plot
ax.set(title="Heart disease and cholestrol levels",
    xlabel="Age",
      ylabel="cholestrol");
#Add a legend
ax.legend(*scatter.legend_elements(),title="Target");

# add a hortizonatal line
ax.axhline(Over_50["chol"].mean(),
          linestyle='--');
```

```python
# we are going to make combine part two dataframe
fig,(ax0,ax1)=plt.subplots(nrows=2,
                          ncols=1,
                          figsize=(10,10),
                           sharex=True)
# Add data to the 'ax0'
scatter=ax0.scatter(x=Over_50["age"],
                   y=Over_50["chol"],
                   c=Over_50["target"],
                   cmap="winter");

#customise the set
ax0.set(title="heart disease and cholestrol level",
       ylabel="cholestrol")
# change the limit of the x and y.
ax0.set_xlim([50,80])
# Add legend
ax0.legend(*scatter.legend_elements(),title="Target");

#Add a line
ax0.axhline(Over_50["chol"].mean(),
            linestyle="--");

#Adding data to the ax1
scatter=ax1.scatter(x=Over_50["age"],
                    y=Over_50["thalach"],
                    c=Over_50["target"],
                   cmap="winter");
ax1.set_xlim([50,80])
ax1.set_ylim([60,200])
ax1.set(title="heart disease and Heart rate",
       xlabel="Age",
       ylabel="Max heart rate");
# Add legend
ax1.legend(*scatter.legend_elements(),title="Target");

#Add a line
ax1.axhline(Over_50["thalach"].mean(),
            linestyle="--");
#Add title to figure
fig.suptitle("Heart Disease Analysis",fontsize=16,fontweight="bold");
```
